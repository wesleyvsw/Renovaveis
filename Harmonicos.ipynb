```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo parâmetros
f = 60  # Frequência em Hz
wm = 2 * np.pi * f  # Pulsação
t = np.linspace(0, 0.1, 1000)  # Tempo de 0 a 0.1 segundos

# Amplitudes
Vm = 1  # Tensão em pu

# Tensões de fase
Va = Vm * np.sin(wm * t)
Vb = Vm * np.sin(wm * t - 2 * np.pi / 3)
Vc = Vm * np.sin(wm * t + 2 * np.pi / 3)

# Plotar as tensões
plt.figure(figsize=(10, 6))
plt.plot(t, Va, label='V_a', color='blue')
plt.plot(t, Vb, label='V_b', color='red')
plt.plot(t, Vc, label='V_c', color='green')
plt.title('Tensões de Fase em um Sistema Trifásico Balanceado')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (pu)')
plt.legend()
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo parâmetros
f = 60  # Frequência em Hz
wm = 2 * np.pi * f  # Pulsação
t = np.linspace(0, 0.1, 1000)  # Tempo de 0 a 0.1 segundos

# Amplitudes
Vm = 1  # Tensão em pu
harmonica = 0.2  # Amplitude da harmônica

# Tensões de fase no referencial estático
Va = Vm * np.sin(wm * t) + harmonica * np.sin(5 * wm * t + np.pi)
Vb = Vm * np.sin(wm * t - 2 * np.pi / 3) + harmonica * np.sin(5 * wm * t - 2 * np.pi / 3 + np.pi)
Vc = Vm * np.sin(wm * t + 2 * np.pi / 3) + harmonica * np.sin(5 * wm * t + 2 * np.pi / 3 + np.pi)

# Tensões no referencial síncrono
Vd = Va  # Tensão direta
Vq = (1 / np.sqrt(3)) * (Vb - Vc)  # Tensão quadrática

# Plotar as tensões no referencial estático
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, Va, label='V_a', color='blue')
plt.plot(t, Vb, label='V_b', color='red')
plt.plot(t, Vc, label='V_c', color='green')
plt.title('Tensões de Fase no Referencial Estático com Harmônicas')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (pu)')
plt.legend()
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')

# Plotar as tensões no referencial síncrono
plt.subplot(2, 1, 2)
plt.plot(t, Vd, label='V_d', color='blue')
plt.plot(t, Vq, label='V_q', color='red')
plt.title('Tensões no Referencial Síncrono com Harmônicas')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (pu)')
plt.legend()
plt.grid()
plt.axhline(0, color='black', lw=0.5, ls='--')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo parâmetros
f = 60  # Frequência em Hz
wm = 2 * np.pi * f  # Pulsação
t = np.linspace(0, 0.1, 1000)  # Tempo de 0 a 0.1 segundos

# Amplitudes
Vm = 1  # Tensão em pu
harmonica = 0.2  # Amplitude da harmônica

# Tensões de fase sem harmônicas
Va_0 = Vm * np.sin(wm * t)
Vb_0 = Vm * np.sin(wm * t - 2 * np.pi / 3)
Vc_0 = Vm * np.sin(wm * t + 2 * np.pi / 3)

# Tensões no referencial síncrono sem harmônicas
Vd_0 = Va_0
Vq_0 = (1 / np.sqrt(3)) * (Vb_0 - Vc_0)

# Tensões de fase com harmônicas
Va_h = Vm * np.sin(wm * t) + harmonica * np.sin(5 * wm * t + np.pi)
Vb_h = Vm * np.sin(wm * t - 2 * np.pi / 3) + harmonica * np.sin(5 * wm * t - 2 * np.pi / 3 + np.pi)
Vc_h = Vm * np.sin(wm * t + 2 * np.pi / 3) + harmonica * np.sin(5 * wm * t + 2 * np.pi / 3 + np.pi)

# Tensões no referencial síncrono com harmônicas
Vd_h = Va_h
Vq_h = (1 / np.sqrt(3)) * (Vb_h - Vc_h)

# Plotar a trajetória do vetor espacial
plt.figure(figsize=(8, 8))
plt.plot(Vd_0, Vq_0, label='Sem Harmônicas', color='blue', alpha=0.5)
plt.plot(Vd_h, Vq_h, label='Com Harmônicas', color='red', alpha=0.5)
plt.title('Trajetória do Vetor Espacial no Referencial Síncrono')
plt.xlabel('V_d (pu)')
plt.ylabel('V_q (pu)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.axis('equal')
plt.show()
```

```python


# Parâmetros do sistema
f = 60  # Frequência fundamental (Hz)
omega = 2 * np.pi * f
t = np.linspace(0, 0.033, 1000)  # Tempo (s)
Vm = 1  # Amplitude da fundamental (pu)
Vh = 0.2 * Vm  # Amplitude da harmônica
theta = 0  # Ângulo inicial da fundamental

# Componentes da tensão
Vf = Vm * np.cos(omega * t + theta)
Vh = Vh * np.cos(5 * omega * t - theta)

# Tensão total
Vt = Vf + Vh

# Plotando os gráficos
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(t, Vf, label='Fundamental')
plt.title('Tensão no Referencial Estático')
plt.ylabel('Tensão (pu)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, Vh, label='Harmônica de 5ª ordem')
plt.ylabel('Tensão (pu)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, Vt, label='Tensão Total')
plt.xlabel('Tempo (s)')
plt.ylabel('Tensão (pu)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
