```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
Vm = 1  # Amplitude da tensão em pu
f = 60  # Frequência em Hz
w = 2 * np.pi * f  # Frequência angular em rad/s
T = 1 / f  # Período do sinal
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tensões trifásicas balanceadas
Va = Vm * np.cos(w * t)  # Tensão fase A
Vb = Vm * np.cos(w * t - 2 * np.pi / 3)  # Tensão fase B (defasada de -120°)
Vc = Vm * np.cos(w * t + 2 * np.pi / 3)  # Tensão fase C (defasada de +120°)

# Transformação para referencial alfa-beta (Clarke)
Valpha = (2/3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2/3) * (np.sqrt(3)/2) * (Vb - Vc)

# Plotagem das tensões trifásicas
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tensões Trifásicas Balanceadas')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes alfa-beta
plt.subplot(2, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tensões no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
Vm = 1  # Amplitude da tensão em pu
f = 60  # Frequência em Hz
w = 2 * np.pi * f  # Frequência angular em rad/s
T = 1 / f  # Período do sinal
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tensões trifásicas balanceadas
Va = Vm * np.cos(w * t)  # Tensão fase A
Vb = Vm * np.cos(w * t - 2 * np.pi / 3)  # Tensão fase B
Vc = Vm * np.cos(w * t + 2 * np.pi / 3)  # Tensão fase C

# Transformação para referencial alfa-beta (Clarke)
Valpha = (2/3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2/3) * (np.sqrt(3)/2) * (Vb - Vc)

# Transformação para referencial síncrono dq (Park)
theta = w * t  # Ângulo de rotação síncrona
Vd = Valpha * np.cos(theta) + Vbeta * np.sin(theta)
Vq = -Valpha * np.sin(theta) + Vbeta * np.cos(theta)

# Plotagem das tensões trifásicas
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tensões Trifásicas Balanceadas')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes alfa-beta
plt.subplot(3, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tensões no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes dq
plt.subplot(3, 1, 3)
plt.plot(t, Vd, label='V_d')
plt.plot(t, Vq, label='V_q')
plt.title('Tensões no Referencial Síncrono (dq)')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do sistema
Vm = 1  # Amplitude fundamental (pu)
Vh = 0.2  # Amplitude da quinta harmônica (pu)
f = 60  # Frequência fundamental (Hz)
w1 = 2 * np.pi * f  # Frequência angular fundamental (rad/s)
w5 = 5 * w1  # Frequência angular da quinta harmônica (rad/s)
T = 1 / f  # Período do sinal fundamental
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tensões trifásicas fundamentais (sequência positiva)
Va1 = Vm * np.cos(w1 * t)
Vb1 = Vm * np.cos(w1 * t - 2 * np.pi / 3)
Vc1 = Vm * np.cos(w1 * t + 2 * np.pi / 3)

# Tensões trifásicas da quinta harmônica (sequência negativa)
Va5 = Vh * np.cos(w5 * t)
Vb5 = Vh * np.cos(w5 * t + 2 * np.pi / 3)
Vc5 = Vh * np.cos(w5 * t - 2 * np.pi / 3)

# Tensões trifásicas resultantes
Va = Va1 + Va5
Vb = Vb1 + Vb5
Vc = Vc1 + Vc5

# Transformação para referencial alfa-beta (Clarke)
Valpha = (2 / 3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2 / 3) * (np.sqrt(3) / 2) * (Vb - Vc)

# Transformação para referencial síncrono dq (Park)
theta = w1 * t  # Ângulo de rotação síncrona
Vd = Valpha * np.cos(theta) + Vbeta * np.sin(theta)
Vq = -Valpha * np.sin(theta) + Vbeta * np.cos(theta)

# Trajetória do vetor espaço (plano alfa-beta)
plt.figure(figsize=(12, 12))

# Gráfico 1: Tensões trifásicas
plt.subplot(4, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tensões Trifásicas com Quinta Harmônica')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Gráfico 2: Referencial Alfa-Beta
plt.subplot(4, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tensões no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Gráfico 3: Referencial DQ
plt.subplot(4, 1, 3)
plt.plot(t, Vd, label='V_d')
plt.plot(t, Vq, label='V_q')
plt.title('Tensões no Referencial Síncrono (dq)')
plt.xlabel('Tempo [s]')
plt.ylabel('Tensão [pu]')
plt.legend()
plt.grid()

# Gráfico 4: Trajetória do vetor espaço
plt.subplot(4, 1, 4)
plt.plot(Valpha, Vbeta, label='Trajetória no Plano Alfa-Beta')
plt.title('Trajetória do Vetor Espaço no Referencial Alfa-Beta')
plt.xlabel('V_alpha [pu]')
plt.ylabel('V_beta [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
