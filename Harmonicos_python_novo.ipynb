```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros do sistema
Vm = 1  # Amplitude da tenso em pu
f = 60  # Frequncia em Hz
w = 2 * np.pi * f  # Frequncia angular em rad/s
T = 1 / f  # Perodo do sinal
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tenses trifsicas balanceadas
Va = Vm * np.cos(w * t)  # Tenso fase A
Vb = Vm * np.cos(w * t - 2 * np.pi / 3)  # Tenso fase B (defasada de -120)
Vc = Vm * np.cos(w * t + 2 * np.pi / 3)  # Tenso fase C (defasada de +120)

# Transformao para referencial alfa-beta (Clarke)
Valpha = (2/3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2/3) * (np.sqrt(3)/2) * (Vb - Vc)

# Plotagem das tenses trifsicas
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tenses Trifsicas Balanceadas')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes alfa-beta
plt.subplot(2, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tenses no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros do sistema
Vm = 1  # Amplitude da tenso em pu
f = 60  # Frequncia em Hz
w = 2 * np.pi * f  # Frequncia angular em rad/s
T = 1 / f  # Perodo do sinal
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tenses trifsicas balanceadas
Va = Vm * np.cos(w * t)  # Tenso fase A
Vb = Vm * np.cos(w * t - 2 * np.pi / 3)  # Tenso fase B
Vc = Vm * np.cos(w * t + 2 * np.pi / 3)  # Tenso fase C

# Transformao para referencial alfa-beta (Clarke)
Valpha = (2/3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2/3) * (np.sqrt(3)/2) * (Vb - Vc)

# Transformao para referencial sncrono dq (Park)
theta = w * t  # ngulo de rotao sncrona
Vd = Valpha * np.cos(theta) + Vbeta * np.sin(theta)
Vq = -Valpha * np.sin(theta) + Vbeta * np.cos(theta)

# Plotagem das tenses trifsicas
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tenses Trifsicas Balanceadas')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes alfa-beta
plt.subplot(3, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tenses no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Plotagem das componentes dq
plt.subplot(3, 1, 3)
plt.plot(t, Vd, label='V_d')
plt.plot(t, Vq, label='V_q')
plt.title('Tenses no Referencial Sncrono (dq)')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros do sistema
Vm = 1  # Amplitude fundamental (pu)
Vh = 0.2  # Amplitude da quinta harmnica (pu)
f = 60  # Frequncia fundamental (Hz)
w1 = 2 * np.pi * f  # Frequncia angular fundamental (rad/s)
w5 = 5 * w1  # Frequncia angular da quinta harmnica (rad/s)
T = 1 / f  # Perodo do sinal fundamental
t = np.linspace(0, T, 1000)  # Vetor de tempo: 1 ciclo com 1000 pontos

# Tenses trifsicas fundamentais (sequncia positiva)
Va1 = Vm * np.cos(w1 * t)
Vb1 = Vm * np.cos(w1 * t - 2 * np.pi / 3)
Vc1 = Vm * np.cos(w1 * t + 2 * np.pi / 3)

# Tenses trifsicas da quinta harmnica (sequncia negativa)
Va5 = Vh * np.cos(w5 * t)
Vb5 = Vh * np.cos(w5 * t + 2 * np.pi / 3)
Vc5 = Vh * np.cos(w5 * t - 2 * np.pi / 3)

# Tenses trifsicas resultantes
Va = Va1 + Va5
Vb = Vb1 + Vb5
Vc = Vc1 + Vc5

# Transformao para referencial alfa-beta (Clarke)
Valpha = (2 / 3) * (Va - 0.5 * Vb - 0.5 * Vc)
Vbeta = (2 / 3) * (np.sqrt(3) / 2) * (Vb - Vc)

# Transformao para referencial sncrono dq (Park)
theta = w1 * t  # ngulo de rotao sncrona
Vd = Valpha * np.cos(theta) + Vbeta * np.sin(theta)
Vq = -Valpha * np.sin(theta) + Vbeta * np.cos(theta)

# Trajetria do vetor espao (plano alfa-beta)
plt.figure(figsize=(12, 12))

# Grfico 1: Tenses trifsicas
plt.subplot(4, 1, 1)
plt.plot(t, Va, label='Va (Fase A)')
plt.plot(t, Vb, label='Vb (Fase B)')
plt.plot(t, Vc, label='Vc (Fase C)')
plt.title('Tenses Trifsicas com Quinta Harmnica')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Grfico 2: Referencial Alfa-Beta
plt.subplot(4, 1, 2)
plt.plot(t, Valpha, label='V_alpha')
plt.plot(t, Vbeta, label='V_beta')
plt.title('Tenses no Referencial Alfa-Beta')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Grfico 3: Referencial DQ
plt.subplot(4, 1, 3)
plt.plot(t, Vd, label='V_d')
plt.plot(t, Vq, label='V_q')
plt.title('Tenses no Referencial Sncrono (dq)')
plt.xlabel('Tempo [s]')
plt.ylabel('Tenso [pu]')
plt.legend()
plt.grid()

# Grfico 4: Trajetria do vetor espao
plt.subplot(4, 1, 4)
plt.plot(Valpha, Vbeta, label='Trajetria no Plano Alfa-Beta')
plt.title('Trajetria do Vetor Espao no Referencial Alfa-Beta')
plt.xlabel('V_alpha [pu]')
plt.ylabel('V_beta [pu]')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
